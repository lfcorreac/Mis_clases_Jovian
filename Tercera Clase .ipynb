# IF , ELSE IF AND ELIF

```python
un_numero=4
```

```python
if un_numero % 2 == 0:
    print("Esta linea esta dentro del bloque if")
    print("El número {} es par".format(un_numero))
```

```python
otro_numero=35
```

```python
if otro_numero % 2 == 0:
    print("El numero {} es par".format(otro_numero))
else:
    print("El numero {} es impar".format(otro_numero))
```

```python
tres_mosqueteros="Luis","Britney","Piero"
```

```python
candidato="Sebastian"
```

```python
if candidato in tres_mosqueteros:
    print("{} es un mosquetero".format(candidato))
else:
    print("{} no es un mosquetero".format(candidato))
```

**Diferencia entre IF Y ELIF**

```python
a_number=15
```

```python
if a_number % 2 == 0:
    print('{} es divisible por 2'.format(a_number))
elif a_number % 3 == 0:
    print('{} es divisible por 3'.format(a_number))
elif a_number % 5 == 0:
    print('{} es divisible por 5'.format(a_number))
elif a_number % 7 == 0:
    print('{} es divisible por 7'.format(a_number))
```

```python
if a_number % 2 == 0:
    print('{} es divisible por 2'.format(a_number))
if a_number % 3 == 0:
    print('{} es divisible por 3'.format(a_number))
if a_number % 5 == 0:
    print('{} es divisible por 5'.format(a_number))
if a_number % 7 == 0:
    print('{} es divisible por 7'.format(a_number))
```

**Para que el** *elif* **se ejecute, es necesario que la condicion previa sea negativa o falsa**

## Usando IF, ELIF Y ELSE JUNTOS

```python
a_number=49
```

```python
if a_number % 2 == 0:
    print('{} es divisible por 2'.format(a_number))
elif a_number % 3 == 0:
    print('{} es divisible por 3'.format(a_number))
elif a_number % 5 == 0:
    print('{} es divisible por 5'.format(a_number))
else:
    print('Todos las sentencias fallaron')
    print('{} no es divisible por 2, 3 o 5'.format(a_number))
```
