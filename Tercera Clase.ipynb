# IF , ELSE IF AND ELIF

- Si se cumple la condición 1, se ejecuta el bloque 1
- Si no se cumple la condición 1 pero sí que se cumple la condición 2, se ejecuta el bloque 2
- Si no se cumplen ni la condición 1 ni la condición 2, se ejecuta el bloque 3.

**SE NOTA QUE:** Solo se pueden declarar condiciones en ***if*** y ***elif***

```python
if un_numero % 2 == 0:
    print("Esta linea esta dentro del bloque if")
    print("El número {} es par".format(un_numero))
```

## DECLARACIONES Y VARIABLES - Formas de declarar una condicion

Un caso de uso frecuente de la instrucción if implica probar una condición y establecer el valor de una variable en función de la condición.

```python
aleatorio_1 = 13

if aleatorio_1 % 2 == 0:
    condicion = 'par'
else:
    condicion = 'impar'

print('El numero {} es {}.'.format(aleatorio_1, condicion))
```

Otra forma de declarar

```python
condicion = "par" if aleatorio_1 %2 ==0 else "impar"
print ("el número {} es {}".format(aleatorio_1,condicion))
```

# ITERACION USANDO WHILE

Intentemos calcular el factorial de 100 usando un ciclo while. El factorial de un número n es el producto (multiplicación) de todos los números del 1 al n, es decir, 1 * 2 * 3 * ... * (n-2) * (n-1) * n.

```python
n=1
factorial=1
m=int(input("Calcular el factorial de "))
while n <= m:
    factorial=n*factorial
    n=n+1
print('The factorial de {} is: {}'.format(m,factorial))
```

Puede verificar cuánto tarda una celda en ejecutarse agregando el comando mágico **%%time** en la parte superior de la celda. Prueba a comprobar cuánto tiempo lleva calcular el factorial de 100, 1000, 10000, 100000, etc.

```python
%%time
n=1
factorial=1
while n <= 100:
    factorial=n*factorial
    n=n+1
print('The factorial of 100 is: {}'.format(factorial))
```
